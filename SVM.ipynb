```python
scaler = StandardScaler()
x =df_train.drop('price_range', axis =1)
y = df_train['price_range']   ##0,1,2,3 classification clusters

scaler.fit(x)
x_transformed = scaler.transform(x)

#transoforming your data into a standard scale 0-1

x_train, x_test,y_train,y_test = train_test_split(x_transformed, y,test_size = 0.3)
```

```python
##SVM

model = SVC()

model.fit(x_train, y_train)

y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)


print('Train Set Accuracy:' +str(accuracy_score(y_train_pred, y_train)*100))
print('Test Set Accuracy:'+ str(accuracy_score(y_test_pred,y_test)*100))
print('\n Confusion Matrix :\n%s'%confusion_matrix(y_test_pred,y_test))
print('\n Classification_Report :\n%s'%classification_report(y_test_pred, y_test))
```
