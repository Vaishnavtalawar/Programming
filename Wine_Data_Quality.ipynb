```python
plt.figure(figsize=[15,6])
plt.bar(df['quality'],df['alcohol'])
plt.xlabel('quality')
plt.ylabel('alcohol')
plt.show()
```

From above graph we can decide, as alcohol value ranges 80-90%,quality is good

```python
from sklearn.linear_model import LinearRegression
```

```python
new_df = df.drop(['quality'], axis =1)

new_df.head()
```

```python
qlt_df= df['quality']
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_df,qlt_df, test_size = 0.2, random_state = 4)

print(qlt_df.shape,y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model =LinearRegression()
model.fit(x_train, y_train)
```

```python
accuracy = model.score(x_test, y_test)
print(accuracy*100,'%')
```
